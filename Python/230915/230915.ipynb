## Q1. 자격증명

### +,-,공백으로 이루어진 문자열을 주면 이를 이진수로 바꾸어 아스키코드와 매치시키기
#### [' + - - + - + - ', ' + + + - + - + ', ' + + - + + + - ']
->  [1001010 , 1110101, 1101110 ]
->['74', '117', '110']
->'Jun'

```python
data = [' + - - + - + - ', ' + + + - + - + ', ' + + - + + + - ']
ans = ''
for i in data:
    x = []
    for j in i:
        # print(x)
        if j == ' ':
            continue
        elif j == '+':
            x.append('1')

        else:
            x.append('0')
    y = ''.join(x)
    z = int(y,2)
    w = chr(z)

    ans+=w

print(ans)
```
